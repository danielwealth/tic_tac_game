```python
import random

def play_game():
    # Define possible choices
    choices = ["rock", "paper", "scissors"]

    play = True
    while play:
        print("THE COMPUTER HAS SELECTED BETWEEN ROCK, PAPER AND SCISSORS. GUESS WHAT THE COMPUTER SELECTED")

        # Get user choice
        user_choice = input("Enter your choice (rock, paper, scissors or 'quit' to exit): ").lower()
        if user_choice == "quit":
            print("Thanks for playing! Goodbye.")
            play = False
            break
        if user_choice not in choices:
            print("Invalid choice. Please select rock, paper, or scissors.")
            continue

        # Generate computer choice
        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")

        # Determine the winner
        if user_choice == computer_choice:
            print("You got it!")
            play= False
        else:
            print('x'*50)
            print("Not quite. You can try again")

# Run the game
play_game()
```

```python
import random

# Define the game board
board = [" " for _ in range(9)]

def print_board():
    print(f"{board[0]} | {board[1]} | {board[2]}")
    print("--+---+--")
    print(f"{board[3]} | {board[4]} | {board[5]}")
    print("--+---+--")
    print(f"{board[6]} | {board[7]} | {board[8]}")

# Check if a player has won
def check_winner(player):
    win_conditions = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # columns
        [0, 4, 8], [2, 4, 6]              # diagonals
    ]
    for condition in win_conditions:
        if all(board[i] == player for i in condition):
            return True
    return False

# Check if the game is a tie
def check_tie():
    return " " not in board

# Simple AI agent: chooses a random empty spot
def ai_move():
    empty_spots = [i for i, spot in enumerate(board) if spot == " "]
    return random.choice(empty_spots)

# Main game loop
def play_game():
    print("Welcome to Tic-Tac-Toe!")
    print_board()

    while True:
        # User move
        user_move = int(input("Enter your move (0-8): "))
        if board[user_move] != " ":
            print("Invalid move. Try again.")
            continue
        board[user_move] = "X"
        print_board()

        if check_winner("X"):
            print("You win!")
            break
        if check_tie():
            print("It's a tie!")
            break

        # AI move
        comp_move = ai_move()
        board[comp_move] = "O"
        print(f"Computer chose position {comp_move}")
        print_board()

        if check_winner("O"):
            print("Computer wins!")
            break
        if check_tie():
            print("It's a tie!")
            break

# Call the main game loop
play_game()
```
